Due to the length of certain headlines approaching 60 tokens, the limit of tokens per headline will be 20.  This will account for certain headlines that include the source at the beginning, such as 'BBC: ' but will also truncate the last part of headlines which sometimes include commentary from the individual who posted the article.

```python
def clean_data_list(element):
    return [w for w in CountVectorizer().build_tokenizer()(str(element).lower()) if w not in swords][:20]
```

```python
unmerged_news_list = testnews.applymap(clean_data_list)
```

### Outputting tokens to string instead

```python
# Returns a list of max 20 lower-cased tokens with stop words removed
# Good for unigrams or TF-IDF, not so much for bigrams or trigrams.

def clean_data_str(element):
    return ' '.join([w for w in CountVectorizer().build_tokenizer()(str(element).lower()) if w not in swords][:20])
```

```python
unmerged_news_str = testnews.applymap(clean_data_str)
```

```python
unmerged_news_str.head()
```

```python
unmerged_news_str.iloc[277,23]
```

```python
unmerged_news_str.to_csv('stocknews//tokens_str.csv', encoding='utf-8', index=False)
```

### Outputting Tokens with Stopwords Included

```python
# Returns a list of max 20 lower-cased tokens with stop words included
# Good for bigrams or trigrams.

def clean_data_str_wsw(element):
    return ' '.join([w for w in CountVectorizer().build_tokenizer()(str(element).lower())][:20])
```

```python
tokens_str_sw = testnews.applymap(clean_data_str_wsw)
```

```python
tokens_str_sw.to_csv('stocknews//tokens_str_sw.csv', encoding='utf-8', index=False)
```

### Outputting Tokens with Stopwords Included and not trimming

```python
def clean_data_str_all(element):
    return ' '.join([w for w in CountVectorizer().build_tokenizer()(str(element).lower())])
```

```python
tokens_str_all = testnews.applymap(clean_data_str_all)
```

```python
tokens_str_all.to_csv('stocknews//tokens_str_all.csv', encoding='utf-8', index=False)
```
